```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

# Path to the input file containing S&P 500 companies
input_file = "/content/drive/MyDrive/S&P_500_Analysis/sp500_companies.csv"

# Output file path for all tickers
output_file_all = "/content/drive/MyDrive/S&P_500_Analysis/sp500_data.csv"

# Read the CSV file and get the tickers
sp500_data = pd.read_csv(input_file)
tickers = sp500_data["Symbol"].dropna().tolist()  # Ensure no NaN values

# Fetch data for all companies

def fetch_company_data(ticker):
    try:
        # Fetch ticker data
        stock = yf.Ticker(ticker)

        # Define the date range
        end_date = datetime.today()
        start_date = end_date - timedelta(days=180)  # 6 months back

        # Fetch historical data
        historical_data = stock.history(start=start_date, end=end_date, interval="1d")

        # Ensure there is data available
        if historical_data.empty:
            print(f"No historical data for {ticker}")
            return None

        # Get the most recent close and the close from 6 months ago
        recent_close = historical_data["Close"].iloc[-1]
        recent_date = historical_data.index[-1].strftime('%Y-%m-%d')
        six_months_close = historical_data["Close"].iloc[0]
        six_months_date = historical_data.index[0].strftime('%Y-%m-%d')

        # Collect metadata
        info = stock.info
        name = info.get("shortName", "N/A")
        sector = info.get("sector", "N/A")
        industry = info.get("industry", "N/A")
        pe_ratio = info.get("trailingPE", "N/A")
        eps = info.get("trailingEps", "N/A")
        dividend_yield = info.get("dividendYield", "N/A")

        # Create a single-row dictionary
        company_data = {
            "Ticker": ticker,
            "Name": name,
            "Sector": sector,
            "Industry": industry,
            "Recent_Close": recent_close,
            "Recent_Date": recent_date,
            "Six_Months_Close": six_months_close,
            "Six_Months_Date": six_months_date,
            "PE_Ratio": pe_ratio,
            "EPS": eps,
            "Dividend_Yield": dividend_yield
        }

        return company_data

    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return None

# Iterate over all tickers and collect data
all_data = []
for ticker in tickers:
    print(f"Processing {ticker}...")
    company_data = fetch_company_data(ticker)
    if company_data:
        all_data.append(company_data)

# Convert the collected data into a DataFrame
all_data_df = pd.DataFrame(all_data)

# Save the aggregated data for all companies
all_data_df.to_csv(output_file_all, index=False)
print(f"Data for all companies saved to {output_file_all}")
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
import time

# Path to the input file containing S&P 500 companies
input_file = "/content/drive/MyDrive/S&P_500_Analysis/sp500_companies.csv"

# Output file path for remaining tickers
output_file_remaining = "/content/drive/MyDrive/S&P_500_Analysis/sp500_data_v2.csv"

# Read the CSV file and get the tickers
sp500_data = pd.read_csv(input_file)
tickers = sp500_data["Symbol"].dropna().tolist()  # Ensure no NaN values

# Fetch data for all companies
def fetch_company_data(ticker):
    try:
        # Fetch ticker data
        stock = yf.Ticker(ticker)

        # Define the date range
        end_date = datetime.today()
        start_date = end_date - timedelta(days=180)  # 6 months back

        # Fetch historical data
        historical_data = stock.history(start=start_date, end=end_date, interval="1d")

        # Ensure there is data available
        if historical_data.empty:
            print(f"No historical data for {ticker}")
            return None

        # Get the most recent close and the close from 6 months ago
        recent_close = historical_data["Close"].iloc[-1]
        recent_date = historical_data.index[-1].strftime('%Y-%m-%d')
        six_months_close = historical_data["Close"].iloc[0]
        six_months_date = historical_data.index[0].strftime('%Y-%m-%d')

        # Collect metadata
        info = stock.info
        name = info.get("shortName", "N/A")
        sector = info.get("sector", "N/A")
        industry = info.get("industry", "N/A")
        pe_ratio = info.get("trailingPE", "N/A")
        eps = info.get("trailingEps", "N/A")
        dividend_yield = info.get("dividendYield", "N/A")

        # Create a single-row dictionary
        company_data = {
            "Ticker": ticker,
            "Name": name,
            "Sector": sector,
            "Industry": industry,
            "Recent_Close": recent_close,
            "Recent_Date": recent_date,
            "Six_Months_Close": six_months_close,
            "Six_Months_Date": six_months_date,
            "PE_Ratio": pe_ratio,
            "EPS": eps,
            "Dividend_Yield": dividend_yield
        }

        return company_data

    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return None

# Iterate over all tickers and collect data
remaining_data = []
batch_size = 50  # Number of tickers to process in one batch
pause_duration = 30  # Pause duration in seconds between batches

for i in range(0, len(tickers), batch_size):
    batch = tickers[i:i + batch_size]
    for ticker in batch:
        print(f"Processing {ticker}...")
        company_data = fetch_company_data(ticker)
        if company_data:
            remaining_data.append(company_data)
    # Pause to avoid throttling
    print(f"Pausing for {pause_duration} seconds to avoid server throttling...")
    time.sleep(pause_duration)

# Convert the collected data into a DataFrame
remaining_data_df = pd.DataFrame(remaining_data)

# Save the aggregated data for all companies
remaining_data_df.to_csv(output_file_remaining, index=False)
print(f"Data for all companies saved to {output_file_remaining}")
```
